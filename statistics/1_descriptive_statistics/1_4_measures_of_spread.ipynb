# Measures of spread

Previously I showed [measures of location](1_2_measures_of_location.ipynb) which describe the central tendency of the data, i. e., through a number we can have an idea about where most of the data is located, they summarize what are the typical elements in a dataset.

In descriptive statistics, on the other side, some measures tell us how spread a dataset is. They are known as measures of spread or measures of dispersion. It's the opposite of seeking to understand the central tendency of the data, here we want to understand how spread the data are: **range**, **IQR**, **mean absolute deviation**, **variance** and **standard deviation**.

But before talking about them I want to introduce some measures that separate or cut the data. I already talked about the median which cuts the data set in half, i. e., 50% of the data to the left and 50% to the right.

Other measures that cut the dataset are:

<ul>
<li><strong>Quartiles:</strong>&nbsp;distribution is divided into quarters</li>
<li><strong>Quintiles:</strong>&nbsp;distribution is divided into fifths.</li>
<li><strong>Deciles:</strong>&nbsp;distribution is divided into tenths.</li>
<li><strong>Percentile:&nbsp;</strong>distribution is divided into hundredths.</li>
</ul>

These values are not influenced by extreme values (outliers) as they just cut or separate the dataset at specific points, they don't depend on the data values, but on the size of the data.

Part of this concept is the fact that if you want to split the dataset into $ n $ equal parts you'll need to $ n - 1 $ dividers, for example, the median divide the dataset into 2 equal parts, therefore we need only 1 divider. 

It's important to remember that I'm talking about a dataset that is ordered.

# Range

The range gives us a measurement of how spread out the entirety of our data set is. It is the difference between the highest and lowest values within a set of numbers. To calculate the range, subtract the smallest number from the largest number in the set. It is affected by extreme values since we consider both ends of a dataset.

The range is sensitive to outliers because it uses the whole dataset:

# $$ \text{range} = x_{(n)} - x_{(1)} $$

# IQR

Stands for Inter Quartile Range and is this blue box you see on the plots above. It shows the spread of the middle half of a distribution, i. e., 50%, the most frequent values of the data are within this range:

# $$ IQR = Q_3 - Q_1 $$

> The interquartile range is an especially useful measure of variability for skewed distributions. For these distributions, the median is the best measure of central tendency because it’s the value exactly in the middle when all values are ordered from low to high. Along with the median, the IQR can give you an overview of where most of your values lie and how clustered they are. The IQR is also useful for datasets with outliers. Because it’s based on the middle half of the distribution, it’s less influenced by extreme values. [Source](https://www.scribbr.com/statistics/interquartile-range/)

It's a measure more resistant to outliers, just like the median is more resistant as a measure of location.

The box plot is also known as box and whisker plot. The name whisker comes from the fact that it has these lines that extend from the box and you might ask what defines where they end. I tell you.

On the upper side you calculate:

# $$ Q_3 + 1.5 \times IQR $$

On the lower side you calculate:

# $$ Q_1 - 1.5 \times IQR $$

Then you get the boundaries for the whiskers. Any values that fall beyond those boundaries are considered outliers. 



```python
income_with_less_outliers = data["Income"][data["Income"] <= 4000]
q1, q3 = income_with_less_outliers.quantile([.25, .75])
income_range = income_with_less_outliers.max() - income_with_less_outliers.min() 
income_iqr = q3 - q1
whisker_upper_bondary = q3 + 1.5 * income_iqr
whisker_lower_bondary = q1 - 1.5 * income_iqr

print(f"The range for the income is: {income_range}")
print(f"The IQR for the income is: {income_iqr}")
print(f"The whisker upper boundary is: {whisker_upper_bondary}")
print(f"The whisker upper boundary is: {whisker_lower_bondary}")

# You can confirm this by checking the box plot.
# When the q1 is negative is stopped at 0. Here it happens because there are a lot of incomes with value 0.
# Maybe the data is corrupted and need to be cleaned, but you get the point.
```

## Mean absolute deviation (MAD)

> The mean absolute deviation of a dataset is the average distance between each data point and the mean. It gives us an idea about the variability in a dataset. [Source](https://www.khanacademy.org/math/statistics-probability/summarizing-quantitative-data/other-measures-of-spread/a/mean-absolute-deviation-mad-review)

# $$ MAD =  \frac{1}{n} \sum_{i=1}^n |x_i - \bar{x}| $$

We take the absolute values when calculate how each value deviates from the mean because otherwise the values would cancel each other the sum would result in 0.

The use of measures of dispersion is to understand the variability of data which measures of location can't tell us. For example, we might have two datasets with the same median, but the variability in the data is higher in one than the other. We need a measure of spread to understand variability.

```python
height_mad = data["Height"].mad()
income_mad = data["Income"].mad()

print(height_mad, income_mad)
print("There is much more variability in the income.")
```

Let's visualize what the mean absolute deviation means. I'll create a simple dataset and plot how each datapoint deviates from the mean.

```python
dataset = pd.Series([3, 5, 7, 8, 9, 10, 6, 7, 5])
mean = np.mean(dataset)

fig, ax = plt.subplots(figsize=(20,10))
dataset.plot(style = "o", ax=ax);

ax.hlines(y=mean, xmin=0, xmax=dataset.shape[0] - 1, colors="r")

for i in range(dataset.shape[0]):
    ax.vlines(x = i, ymin=mean, ymax=dataset[i], linestyles="dashed")

```

The mean absolute deviation is the average value of the length of the dashed lines, i. e., how far each datapoint is from the mean.

## Variance

The calculation is very similar to the mean standard deviation, but instead of taking the absolute values of the deviation of each point from the mean, we square those differences, again this is to avoid getting 0 as the sum of the deviations.

- Find the mean of the data set
- Subtract the mean from each observation to find how far each observation is from the mean, i. e., what's the deviation from the mean.
- Square that deviation
- Sum up all deviations for all the data points, i. e., it's the sum of squared deviations.
- Divide by the sample size minus 1 if it's a sample or for the same size if the data is the whole population.
- The more spread the data, the higher the variance.


### Sample

## $$ s^2 = \frac{1}{n-1} \sum^n_{i=1} (x_i - \bar{x})^2 $$

### Population

## $$ \sigma^2 = \frac{1}{N} \sum^N_{i=1} (X_i - \mu)^2 $$

The numerator of this formula is called the sum of squared deviations or corrected sum of squares and is represented by $ S_{xx} $.

## $$ s^2 = \frac{S_{xx}}{n-1} $$



```python
height_variance = data["Height"].var(ddof=1)
income_variance = data["Income"].var(ddof=1)

print(height_variance, income_variance)
print("There is much more variability in the income.")
```

## An alternative formula

If you have a large dataset it might be boring to calculate each deviation. Another option you have is to use the following formula:

### Sample

## $$ s^2 = \frac{\sum^n_{i=1} x_i^2 - n\bar{x}^2}{n - 1} $$

### Population

## $$ \sigma^2 = \frac{\sum^N_{i=1} X_i^2 - N\mu^2}{N} $$


```python
alternative_formula = (data["Height"].pow(2).sum() - data["Height"].shape[0] * data["Height"].mean() ** 2) / (data["Height"].shape[0] - 1)
alternative_formula
```

You might have noticed that to calculate the variance we had to square the deviations which means that its value is not in the same units as the data. For example, if the height is measured in meters, now the variance is in meters squared. It's a bit difficult to interpret that. Yeah, you're right. Time for the last measure of spread planned for this notebook:

## Standard deviation

Since we squared the deviations to calculate the variance, the variance is not in the same unit as the dataset anymore. By taking the square root of the variance we bring the spread of the data back to the dataset unit. It's the square root of the variance.

- Find the variance of the data set

- Take the positive square root of the variance


### Sample

## $$ s^2 = \sqrt{\frac{\sum^n_{i=1} x_i^2 - n\bar{x}}{n - 1}} $$

### Population

## $$ \sigma^2 = \sqrt{\frac{\sum^N_{i=1} X_i^2 - n\mu}{N}} $$

Disclaimer: here I am using the alternative formula for the variance and just taking the square root, but you can take the more "traditional" formula above as well. The key takeaway here is that the standard deviation is just the square root of the variance because we want to undo the step where we squared the distances from the mean and bring the value to the same unit we have in the dataset.

```python
height_std = data["Height"].std(ddof=1)
income_std = data["Income"].std(ddof=1)

print(height_std, income_std)
print("There is much more variability in the income.")
```

### 🤔 Why $ n - 1 $ for samples?

We have a more precise estimation of the population variance. Or getting an unbiased estimation of the population variance.


**Parameter**: when we calculate a measure from the population

**Statistic**: when we calculate a measure from a sample

When we divide by a smaller number (n - 1) we get a larger value. We refer to it as an unbiased estimate.

Why dividing by n only is a biased estimation? Why the unbiased variance gives us a better estimation of the true population variance?

If we divide only by n we underestimate the population variance.

Becaudse the differences betwwen the data points and sample mean tend to be smaller. therefere o larger variance.

## Visualization

- Points on a line is a population N = 14. Imagine the mean of the population.
- we take a sample n = 3. 
- When we sample we can take points close from each other we don't mirror the actual population mean. We can have a lower estimate.
- It's possible that we take a sample like that and the variance will be wrong. We migh understimate the true population variance.
- We divide by n-1 so we get a larger population variance.
- When the sample mean is far from the true population mean the variance will also be wrongly estimated.
- bigger sample sizer improve the chanches of getting more diverse data and get a more accurate estimate. The chances of getting points close to each other decrease.

When we use the biased estimate we are not appoaching the population variance, but approaching n-1. Try to reproduce khan academy chart = https://www.khanacademy.org/math/ap-statistics/summarizing-quantitative-data-ap/more-standard-deviation/v/simulation-showing-bias-in-sample-variance

- so, if we want to get an unbiased estimate we would multiply the variance by n / (n-1) to cancel the bias.

```python
population = data["Age"]
print(f"The population mean (parameter) is {round(np.mean(population), 4)}")
print(f"The population variance (parameter) is {round(np.var(population), 4)}")

sample = np.random.choice(data["Age"], 100)
print(f"The sample mean (statistic) is {round(np.mean(sample), 4)}")
print(f"The biased sample variance estimate (statistic) is {round(np.var(sample), 4)}")
```

Above you will see that the biased estimation of the population variance, i. e., divided by $ n $ and not $ n - 1 $ is underestimated, this means that the biased sample variance is lower than the actual sample variance. Let's move to the unbiased variance estimate, i. e., divide the sum of the squared distances from the mean by $ n - 1 $ and we'll move to understand why it works.

```python
population = data["Age"]
print(f"The population mean (parameter) is {round(np.mean(population), 4)}")
print(f"The population variance (parameter) is {round(np.var(population), 4)}")

sample = np.random.choice(data["Age"], 100)
print(f"The sample mean (statistic) is {round(np.mean(sample), 4)}")
print(f"The unbiased sample variance estimate (statistic) is {round(np.var(sample, ddof=1), 4)}")
```

One important thing to note here: when we divide by $ n - 1 $ we're dividing by a smaller number and therefore the variance value that would be underestimated is corrected because we get a larger value when you divide by a smaller number. Think about it: if you have a pizza with 8 slices and need to divide that among 4 people, each person gets 2 slices. If instead, you need to divide the pizza between 2 people each one gets 4 slices.

Sampling is a random process which means that depending on the sample you draw it can be that the sample mean is close to the population mean, but there is a possibility that by chance the data points drawn are far from the actual population mean and this would result in an underestimated statistic.

Below I'm gonna use the height variable to visualize the underestimation and overestimation of the population variance when using the biased variance calculation from a sample. Just for now, I'll consider all the height measurements I have in the dataset as the population and will sample from it.

```python
fix, ax = plt.subplots(1,1, figsize=(8, 5));
pop_mean, pop_size, pop_variance = round(np.mean(data['Height']), 4), data['Height'].shape[0], round(np.var(data['Height']), 6)
plt.suptitle("Histogram: Frequency distribution for the Height variable", fontsize=16)
ax.set_title(f"N: {pop_size}, Mean: {pop_mean}, Variance: {pop_variance}", fontsize=12);
ax.set_xlabel("Height", fontsize=16);
sns.histplot(data["Height"], ax=ax, bins=20);
```

```python
def calculate_statistics_from_sample(sample):
    """
    Returns a tuple with the sample mean, the sample BIASED variance and the sample size.
    """

    return np.mean(sample), np.var(sample), sample.size

n_samples = 10000
samples_statistics = [calculate_statistics_from_sample(np.random.choice(data["Height"], np.random.choice(range(2, 21, 1)))) for i in range(n_samples)]
fix, ax = plt.subplots(1,1, figsize=(8, 5));
pop_mean, pop_size, pop_variance = round(np.mean(data['Height']), 4), data['Height'].shape[0], round(np.var(data['Height']), 6)

samples_statistics_transposed = np.array(samples_statistics)
samples_statistics_transposed = samples_statistics_transposed.transpose()
x, y, size = samples_statistics_transposed[0], samples_statistics_transposed[1], samples_statistics_transposed[2]

plt.suptitle(f"Number of samples: {n_samples}", fontsize=16)
ax.set_title(f"N: {pop_size}, Mean: {pop_mean}, Variance: {pop_variance}", fontsize=12);
ax.set_xlabel("Sample means", fontsize=16);
ax.set_ylabel("Sample variances (biased)", fontsize=16);
plt.axvline(x=pop_mean, linewidth=2, color="r", linestyle="--");
plt.axhline(y=pop_variance, linewidth=2, color="r", linestyle="--");
sns.scatterplot(x=x, y=y, ax=ax, c=size);
```

What you can see on the plot above is that in the samples which had a mean closer to the actual population mean (the vertical red line) the biased sample variances tend to be closer to the population variance. When this doesn't happen the sample variances are either underestimated or overestimated.

So one conclusion here is that the value of the mean will affect the value of the variance because in the formula to calculate the variance we need the mean.

Another conclusion is that in the plot above the points with darker color have a smaller sample size, which you might conclude is that it's not representative of the population so the mean and therefore the variance will be wrongly estimated. The sample size is not the only thing that affects here. You will notice that some lighter colors are being overestimated. The reason is that sampling happens by chance so even if you get a larger sample you can still get random data points whose means are far from the actual population mean.

```python
df = pd.DataFrame(
    data={
        "Sample mean": x,
        "Sample variance (biased)": y,
         "Sample variance (biased) / Population variance": y / pop_variance,
        "Sample size": size.astype(int),
    }
)
df
```

```python
statistics_grouped_by_sample_size = df.groupby("Sample size").mean()
statistics_grouped_by_sample_size[
    "Sample size"
] = statistics_grouped_by_sample_size.index


fix, ax = plt.subplots(1, 1, figsize=(8, 5))
plt.suptitle(f"Number of samples: {n_samples}", fontsize=16)
ax.set_title(f"N: {pop_size}, Mean: {pop_mean}, Variance: {pop_variance}", fontsize=12)
sns.barplot(
    data=statistics_grouped_by_sample_size,
    x="Sample size",
    y="Sample variance (biased) / Population variance",
    ax=ax,
)
```

In the plot above for every sample size, I divide each sample's biased variance by the true population variance. If they were equal,  i. e., the estimate was precise we'd get 1 and the precision would be 100%. Since we already know this is not the case I'm doing this operation for every sample of a specific sample size. I take the mean over all those sample sizes so if all the biased estimates across all the samples of a specific size would be precise we should get a value of 1.

But...

If you observe the plot above, look at sample size 2. We don't get 1, but we get a value close to 0.5 and what this tells us is that we don't get the actual population variance, but a half. Similar to sample size three we get 2/3 and so on... If we generalize this, when we calculate the biased variances we dont get the actual population variance $ \sigma^2 $, but instead what we get is:

# $$ \frac{n-1 }{n} \sigma^2 $$

For example, in the case of sample size 2: 

# $$ \frac{2-1 }{2} \sigma^2 = 0.5 \sigma^2  $$

For example, in the case of sample size 3: 

# $$ \frac{3-1 }{3} \sigma^2 = 0.6667 \sigma^2  $$

So all of this is giving us a biased estimate of the population variance. How ho we unbias it? We have to cancel that bias term somehow so we're left only with the actual population variance:


# $$ \frac{n-1 }{n} \sigma^2 \rightarrow \frac{n-1 }{n} \sigma^2 \times \frac{n }{n-1} $$

But now instead of using the symbol for the variance, let's bring it's formula in and apply the same process:

## $$ s^2 = \frac {\sum^n_{i=1} (x_i - \bar{x})^2}{n} \times \frac{n }{n-1} = \frac {\sum^n_{i=1} (x_i - \bar{x})^2}{n-1}  $$

So here we are back to the variance formula for a sample I showed earlier. Aka the unbiased variance.
What I want to do now is to repeat all the plots I showed up to now, but use the unbiased variance now to see how it gets closer to the actual population parameter.


```python
def calculate_statistics_from_sample(sample):
    """
    Returns a tuple with the sample mean, the sample UNBIASED variance and the sample size.
    """

    return np.mean(sample), np.var(sample, ddof=1), sample.size

n_samples = 10000
samples_statistics = [calculate_statistics_from_sample(np.random.choice(data["Height"], np.random.choice(range(2, 21, 1)))) for i in range(n_samples)]
fix, ax = plt.subplots(1,1, figsize=(8, 5));
pop_mean, pop_size, pop_variance = round(np.mean(data['Height']), 4), data['Height'].shape[0], round(np.var(data['Height']), 6)

samples_statistics_transposed = np.array(samples_statistics)
samples_statistics_transposed = samples_statistics_transposed.transpose()
x, y, size = samples_statistics_transposed[0], samples_statistics_transposed[1], samples_statistics_transposed[2]

plt.suptitle(f"Number of samples: {n_samples}", fontsize=16)
ax.set_title(f"N: {pop_size}, Mean: {pop_mean}, Variance: {pop_variance}", fontsize=12);
ax.set_xlabel("Sample means", fontsize=16);
ax.set_ylabel("Sample variances (unbiased)", fontsize=16);
plt.axvline(x=pop_mean, linewidth=2, color="r", linestyle="--");
plt.axhline(y=pop_variance, linewidth=2, color="r", linestyle="--");
sns.scatterplot(x=x, y=y, ax=ax, c=size);
```

#### Now on to the maths...

The first thing I'll do is use a simple dataset and fix its mean on a value. That fixed value is not the actual mean value from that dataset and from there I'll calculate the dataset variance and will plot that relationship.


```python
population = data["Height"]
mean = np.mean(population)
variance = np.var(population, ddof=1)
print(f"The real mean for the population is {mean}")
print(f"The real variance for the population is {variance}")


def calculate_biased_variance(sample, fixed_mean):
    sum_of_squared_distances = np.power(sample - fixed_mean, 2).sum()
    biased_variance = sum_of_squared_distances / sample.size
    return biased_variance


fixed_means = np.linspace(1.4, 1.9)

sample1 = np.random.choice(population, 100)
sample2 = np.random.choice(population, 100)
sample3 = np.random.choice(population, 100)

biased_variances_1 = np.array(
    [calculate_biased_variance(sample1, x) for x in fixed_means]
)

biased_variances_2 = np.array(
    [calculate_biased_variance(sample2, x) for x in fixed_means]
)

biased_variances_3 = np.array(
    [calculate_biased_variance(sample3, x) for x in fixed_means]
)

fix, ax = plt.subplots(1, 1, figsize=(8, 5))
plt.suptitle("Fixed means x Biased Variance (Height Variable)", fontsize=16)
ax.set_title(f"Mean: {mean}, Variance: {variance}", fontsize=12)
ax.set_xlabel("Fixed means", fontsize=14)
ax.set_ylabel("Biased variance", fontsize=14)
plt.plot(
    fixed_means,
    biased_variances_1,
    color="r",
    label="Biased variance calculated with fixed mean: Sample 1",
)
plt.plot(
    fixed_means,
    biased_variances_2,
    color="g",
    label="Biased variance calculated with fixed mean: Sample 2",
)
plt.plot(
    fixed_means,
    biased_variances_3,
    color="b",
    label="Biased variance calculated with fixed mean: Sample 3",
)
plt.axvline(x=mean, linewidth=1, color="black", linestyle="--", label="Actual mean")
plt.axhline(
    y=variance, linewidth=1, color="purple", linestyle="--", label="Actual variance"
)

plt.legend(loc="upper left")
plt.grid(True)
```

What you can see on the plat above is that when the sample mean gets closer to the actual population mean the biased variance also gets closer to the actual variance. If we would use the population mean we'd get closer to the actual population variance, but the thing is that these are samples and you can see that even when the mean gets closer to the actual mean the biased variance is still underestimating the actual variance. The key takeaway here: using the sample mean to calculate the variance and dividing by $ n $ only underestimates the population variance.

I used fixed means to demonstrate how the graph would look like, let's use the biased variance formula and set the fixed mean I used which in the formula is the sample mean $ \bar{x} $ and use it as the variable $ b $:

# $$ \text{Biased variance} = \frac{\sum (x_i - b)^2}{n} $$

In the plot above I replaced $ b $ in this formula with the fixed means and plotted the output: the biased variance for that specific fixed mean.

What I'm gonna do now is to take the derivative of the biased variance function with respect to $ v $:

$$ \frac{d}{db} \frac{\sum (x_i - b)^2}{n} $$

What we want to do after finding the derivative is to find the point where the derivative equals 0, i. e., the value of $ b $ which represents the sample mean that outputs the smaller variance.

### Using the chain rule


### $$ \frac{d}{db} =  \frac{ -2}{n} \sum(x_i - b) $$

- Take the derivative of the outer function: $ (g)^2 $
- Multiply by the derivative of the inner function $ g $: $ -1 $. This result derives from the fact that the derivative of $ x_i $ with respect to $ b $ is 0 and the derivative of $ -b $ with respect to $ b $ is -1.



Now let's say we have a sample of $ [x_1, x_2, x_3, x_4] $ (n = 4). And we want to find the value of $ b $ that yields the smaller variance, i. e., where the derivative equals 0: 

$$ \frac{-2}{4} [(x_1 - b) + (x_2 - b) + (x_3 - b) + (x_4 - b)] = 0 $$

Multiplying each side to $ \frac{4}{-2} $ to simplify the expression yields:

$$ \frac{-2}{4} \times \frac{4}{-2}  [(x_1 - b) + (x_2 - b) + (x_3 - b) + (x_4 - b)] = 0 \times \frac{4}{-2} $$

$$ x_1 - b + x_2 - b + x_3 - b + x_4 - b = 0 $$

$$ (x_1 + x_2 + x_3 + x_4) - (b + b + b + b) = 0 $$

$$ (x_1 + x_2 + x_3 + x_4) - 4b = 0 $$

$$  - 4b = - (x_1 + x_2 + x_3 + x_4) $$

$$  b = \frac {x_1 + x_2 + x_3 + x_4}{4} = \bar{x} $$

Which is exactly the mean of the dataset. So the variance is minimized when the value for $ b $ is exactly the value of the mean.

The distances between the data points and the sample mean tend to be smaller. If you would use the population mean the distances would be larger. So this all means that when we use the sample mean to try to estimate the population variance we are underestimating it (unless the sample mean is the same as the population mean which is unlikely to happen).


### How to interpret the standard deviation?

Standard deviation tells how much the data varies from the average. A low standard deviation tells that the data is clustered around the mean. A higher standard deviation tells the data is dispersed over a wider range of values.

The standard deviation is useful to understand if a specific data point is STANDARD and EXPECTED or UNUSUAL and UNEXPECTED.

In real life datasets will vary, the standard variation is a measure that helps us to conduct statistical tests and understand if a specific result is statistcally signication or part of expected variation.

#### 68–95–99.7 rule

> In statistics, the 68–95–99.7 rule, also known as the empirical rule, is a shorthand used to remember the percentage of values that lie within an interval estimate in a normal distribution: 68%, 95%, and 99.7% of the values lie within one, two, and three standard deviations of the mean, respectively.

> In mathematical notation, these facts can be expressed as follows, where $ Pr() $ is the probability function, Χ is an observation from a normally distributed random variable, μ (mu) is the mean of the distribution, and σ (sigma) is its standard deviation:

$$ {\displaystyle {\begin{aligned}\Pr(\mu -1\sigma \leq X\leq \mu +1\sigma )&\approx 68.27\%\\\Pr(\mu -2\sigma \leq X\leq \mu +2\sigma )&\approx 95.45\%\\\Pr(\mu -3\sigma \leq X\leq \mu +3\sigma )&\approx 99.73\%\end{aligned}}}
 $$

### Should I use the Mean Absolute Deviation (MAD) or the Standard Deviation (STD)?

**A mathematical approach**

When I took the derivative of the biased variance to show that the biased variance always underestimates the population variance it was possible because that function was differentiable at every point, especially at its minima. To optimize the mean absolute deviation formula we'll take its derivative to find the lowest variance, but at the lowest point, the function is not differentiable because it has a cusp and we can't calculate its slope.

When we move to inferential statistics we'll use calculus to find optimal estimators, i. e., likelihood functions. It's only possible to different such functions if they are continuous.

**An intuition approach**

The MAD measures how far each observation individually is from the mean of all observations, but it doesn't tell how the observations are arranged about one another. Since we square the differences to calculate the standard deviation you can derive that the standard deviation will always be equal to or larger than the mean absolute deviation. This means that when outliers are present in the dataset the standard deviation will be considerably larger than the mean absolute deviation. So the standard deviation gives us the average of deviations, but also it takes into consideration the arrangement of a distribution.

That's why the standard deviation is preferred over the mean absolute deviation in most cases.
